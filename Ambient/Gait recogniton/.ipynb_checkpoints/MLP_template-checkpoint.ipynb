```python
import tensorflow as tf
from python_speech_features import mfcc
import wavio
import numpy as np
import os
from sklearn.cluster import KMeans
tf.reset_default_graph()
sess = tf.InteractiveSession()
```

```python
#input preprocessing
num_classes = 2; input_size = 26
n_train = 4500; n_test = 4500
train_x = []; test_x = []
train_t = []; test_t = []
for i in range(num_classes):
    temp_data = wavio.read('at1/d'+str(i)+'.WAV').data
    l = mfcc(temp_data[:,0],44100,0.5)
    r = mfcc(temp_data[:,1],44100,0.5)
    fx = np.concatenate((l,r),1)
    for j in range(n_train+n_test):
        if j%(2) == 0:
            train_x.append(fx[j])
            train_t.append(i)
        else:
            test_x.append(fx[j])
            test_t.append(i)
train_x = np.array(train_x); train_t = np.array(train_t)
test_x = np.array(test_x); test_t = np.array(test_t)
```

```python
#plotting mfcc
import matplotlib.pyplot as plt
plt.plot(fx.reshape(-1,1)[1::26])
plt.show()
# import matplotlib.pyplot as plt
# plt.imshow(fx[:150].T, cmap='hot', interpolation='nearest')
# plt.show()
```

```python
#hyperparameters
hidden_shape = [100,75,50,23]
act_fns = ['relu','relu','relu','relu']
net_shape = [input_size] + hidden_shape + [num_classes]
num_lyr = len(hidden_shape)
```

```python
learning_rate = tf.placeholder(tf.float32,[])
x = tf.placeholder(tf.float32, [None, input_size])
labels = tf.placeholder(tf.int32, [None])
keep_prob = tf.placeholder(tf.float32,[num_lyr+1])
```

```python
#parameters
params = {'W':{},'b':{}}
for lay_num in range(num_lyr+1):
    params['W'][lay_num] = tf.Variable(tf.random_normal([net_shape[lay_num], net_shape[lay_num+1]],stddev = 0.1))
    params['b'][lay_num] = tf.Variable(tf.random_normal([net_shape[lay_num+1]],stddev = 0.1))
```

```python
#network graph
def act_fn(a,lay_num):
    if act_fns[lay_num] == 'tanh':
        return tf.tanh(a)
    elif act_fns[lay_num] == 'relu':
        return tf.nn.relu(a)
    elif act_fns[lay_num] == 'elu':
        return tf.nn.elu(a)
    
def MLP(x,params):
    z = x
    for lay_num in range(num_lyr):
        z = act_fn(tf.add(tf.matmul(z,params['W'][lay_num]),params['b'][lay_num]),lay_num)
        z = tf.nn.dropout(z,keep_prob[lay_num])
    logits = tf.add(tf.matmul(z,params['W'][lay_num+1]),params['b'][lay_num+1])
    return tf.nn.dropout(logits,keep_prob[lay_num+1])
```

```python
logits = MLP(x,params)
cost = tf.reduce_mean(tf.nn.sparse_softmax_cross_entropy_with_logits(logits,labels))
optimizer = tf.train.GradientDescentOptimizer(learning_rate).minimize(cost)
```

```python
sess.run(tf.initialize_all_variables())
```

```python
n_iter = 0
while True:
    _,cost_ = sess.run([optimizer,cost],feed_dict={x:train_x,labels:train_t,learning_rate:0.1,keep_prob:[0.5]*5})
    if n_iter % 100 == 0:
        print('iter'+str(n_iter)+' cost:'+str(cost_))
        train_logits = sess.run(logits,feed_dict={x:train_x,keep_prob:[1.0]*5})
        print 'train pf = ' + str((sum(np.argmax(train_logits,axis=1) == train_t) + 0.0)/(n_train*num_classes))
        test_logits = sess.run(logits,feed_dict={x:test_x,keep_prob:[1.0]*5})
        print 'test pf = ' + str((sum(np.argmax(test_logits,axis=1) == test_t) + 0.0)/(n_test*num_classes))+'\n'
    n_iter += 1
```

```python
#new input clustering begins
num_classes = 2; input_size = 26
n_size = 8000;
new_x = []; new_t = []
for i in range(num_classes):
    temp_data = wavio.read('at2/d'+str(i)+'.WAV').data
    l = mfcc(temp_data[:,0],44100,0.5)
    r = mfcc(temp_data[:,1],44100,0.5)
    fx = np.concatenate((l,r),1)
    for j in range(n_size):
        new_x.append(fx[j])
        new_t.append(i)
new_x = np.array(new_x); new_t = np.array(new_t)
```

```python
num_lyr = 4
def MLP2(x,params):
    z = x
    for lay_num in range(num_lyr):
        z = act_fn(tf.add(tf.matmul(z,params['W'][lay_num]),params['b'][lay_num]),lay_num)
    return z
    
features = MLP2(x,params)
tfm_data = sess.run(features,feed_dict={x:new_x,keep_prob:[1.0]*5})
pred = KMeans(n_clusters=2).fit_predict(tfm_data)
```
